```python
import pandas as pd
from matplotlib.ticker import FuncFormatter
import seaborn as sns
from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv('final.csv')
```

```python
df.drop(columns = ['Unnamed: 0.1','Unnamed: 0'],inplace=True)
```

```python
df_vacc = df.groupby('location').agg({
    'people_fully_vaccinated':'max'
})
```

```python
df_vacc.sort_values(by = 'people_fully_vaccinated',ascending=False,inplace=True)
```

```python
df_vacc
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data = df_vacc,x = 'location',y = 'people_fully_vaccinated',hue = 'people_fully_vaccinated',palette='flare')
plt.suptitle('Number of People Fully Vaccinated in Each Country',fontsize = 20,fontweight = 'bold',color = 'gray')
sns.set_style('darkgrid')
plt.ylabel('Number of People')
plt.xlabel('Name of the Country')
formatter = plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x,_:f'{int(x):,}'))
plt.legend().set_visible(False)
plt.xticks(rotation = 75)

plt.show()
```
